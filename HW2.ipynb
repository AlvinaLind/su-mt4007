```python
data["ppsqm"] = data["listPrice"]/data["livingArea"]
data[['listPrice', 'livingArea', 'ppsqm']].head()
```

```python
top_5_expensive = data.sort_values(by="ppsqm", ascending=False).head(5)
top_5_expensive[['listPrice', 'livingArea', 'ppsqm']]
```

```python
mean_ppsqm = data["ppsqm"].mean()
print("The average price per square meter is ", mean_ppsqm, "SEK")
```
