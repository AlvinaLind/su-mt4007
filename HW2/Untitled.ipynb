## Omkomna

```python
omkomna_data = pd.read_excel(file_path, sheet_name = "Omkomna", skiprows = 7)

print("Rådata från fliken 'Omkomna':")
omkomna_data.head()
```

```python
omkomna_data = omkomna_data.dropna(how="all", axis=0)  # Ta bort tomma rader
omkomna_data = omkomna_data.dropna(how="all", axis=1)  # Ta bort tomma kolumner

omkomna_data.columns = ["År", "Jan", "Feb", "Mar", "Apr", "Maj", "Jun", 
                        "Jul", "Aug", "Sep", "Okt", "Nov", "Dec", "Summa total"]

# Ta bort rader där "År" inte är numerisk
omkomna_data["År"] = pd.to_numeric(omkomna_data["År"], errors="coerce")
omkomna_data = omkomna_data.dropna(subset=["År"])

omkomna_data.head()
```
